```python
play = df.copy()
```

```python
play = play[['Date','Location','Team','OppTeam','TeamId','GameId','PercentBet_ML','Average_Line_ML','PercentBet_Spread','Average_Line_Spread','Average_Odds_Spread','PercentBet_OU','Average_Line_OU','Average_Odds_OU','Pts','Spread','Result','Total']]
```

```python
play
```

```python
play.dtypes
```

```python
num_rows = len(play.index)
num_cols = len(play.columns)
```

```python
spread_wl = []

for i in np.arange(num_rows):
    given_spread = play.iloc[i,9]
    team_pts = play.iloc[i,14]
    tot_pts = play.iloc[i,17]
    opp_pts = tot_pts - team_pts
    
    if (given_spread + team_pts) > opp_pts:
        spread_wl.append(1)
    else:
        spread_wl.append(0)
        
    
        
```

```python
OU_wl = []

for i in np.arange(num_rows):
    given_OU = play.iloc[i,12]
    actual_OU = play.iloc[i,17]
    
    if (actual_OU > given_OU):
        OU_wl.append(1)
    else:
        OU_wl.append(0)
```

```python
ml_wl = []

for i in np.arange(num_rows):
    result = play.iloc[i,16]
    
    if result == 'W':
        ml_wl.append(1)
    else:
        ml_wl.append(0)
```

```python
play['Spread WL'] = spread_wl
play['Over/Under WL'] = OU_wl
play['MoneyLine WL'] = ml_wl
play
```
